```python
df.replace(" ?", np.nan, inplace=True)
print(df.isnull().sum())
```

###Handling the missing values

```python
df.dropna(inplace=True)
```

```python
# Numerical features → Median
num_cols = df.select_dtypes(include=['int64','float64']).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical features → Mode
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

###Scaling Techniques

####A)Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standard = df.copy()
df_standard[num_cols] = scaler.fit_transform(df[num_cols])
```

####B)Min-Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = minmax.fit_transform(df[num_cols])
```

###Encoding Techniques

####Separate Categorical columns

```python
cat_cols = df.select_dtypes(include='object').columns
```

####A)One-Hot Encoding (<5 categories)

```python
small_cat = [col for col in cat_cols if df[col].nunique() < 5]

df = pd.get_dummies(df, columns=small_cat, drop_first=True)
```

###B)Label Encoding(>5 categories)

```python
from sklearn.preprocessing import LabelEncoder
cat_cols = df.select_dtypes(include='object').columns

large_cat = [col for col in cat_cols if df[col].nunique() >= 5]

le = LabelEncoder()

for col in large_cat:
    df[col] = le.fit_transform(df[col])
```

## Feature Engineering
### Create 2 New Features
####Capital Total Feature

```python
df['total_capital'] = df['capital_gain'] + df['capital_loss']
```

####Education Level Feature

```python
education_order = {
    'Preschool':1, '1st-4th':2, '5th-6th':3, '7th-8th':4,
    '9th':5, '10th':6, '11th':7, '12th':8,
    'HS-grad':9, 'Some-college':10,
    'Assoc-voc':11, 'Assoc-acdm':12,
    'Bachelors':13, 'Masters':14,
    'Prof-school':15, 'Doctorate':16
}

df['education_level'] = df['education'].map(education_order)
```

####Skewed Feature Transformation

```python
print(df[num_cols].skew())
```

```python
#By applying Log Transformation
df['capital_gain_log'] = np.log1p(df['capital_gain'])
```

# Feature Engineering Explanation

1. Total Capital Feature:
The new feature 'total_capital' is created by combining 'capital_gain' and 'capital_loss'.

Rationale:
- Instead of analyzing gain and loss separately, combining them provides a better representation of overall financial status.
- This helps the model understand the net financial impact more effectively.

2. Education Level Feature:
The 'education_level' feature converts categorical education values into ordered numerical values.

Rationale:
- Education has a natural order (e.g., Preschool < Bachelors < Doctorate).
- Converting it into numerical form helps the model capture this hierarchy.
- It improves model performance compared to treating it as simple categories.


##Iosation Forest (OutlierDetection)

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(contamination=0.05)
outliers = iso.fit_predict(df[num_cols])

df['outlier'] = outliers
df = df[df['outlier'] == 1]  # Keep non-outliers
```
